```python
#replace 'top'(3rd row of descibe()) values in place of '?'
columns=['native-country','occupation','workclass']
for col in columns:
    dataset[col].replace([' ?'],[dataset.describe(include='all')[col][2]],inplace=True)
```

```python
dataset['workclass'].value_counts()
#or we can do #print(sum(dataset['native-country']=='?'))
```

```python
from sklearn import preprocessing
le=preprocessing.LabelEncoder()
workplace_tran=le.fit_transform(dataset.education)
dataset['income']=workplace_tran # do this for all the col which are not in interger data i.e after describe() the col which is not shown
```

```python
dataset.describe()
```

```python
x=dataset.drop('income',axis=1)#x is i/p features as we drop the target col
y=dataset['income']#y is o/p as here only the target col
validation_size = 0.20
seed = 7
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=validation_size, random_state=seed)
```

```python
clf=GaussianNB()
clf=clf.fit(X_train, Y_train)
predictions=(clf.predict(X_test))
predictions
```

```python
from sklearn.metrics import accuracy_score , confusion_matrix , classification_report
print(accuracy_score(Y_test,predictions))
```
